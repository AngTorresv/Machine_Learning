# Practica para usar graficos interactivos 

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('population_total-graficos interactivos.csv')
df
```

```python
import cufflinks as cf
from IPython.display import display,HTML 
cf.set_config_file(sharing='public',theme='space',offline=True)

```

```python
cf.getThemes()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df = df.pivot(index='year', columns= 'country',values='population')
df
```

```python
df = df[['United States','India','China','Indonesia','Brazil']]
df
```

# LinePlot


```python
df.iplot(kind='line',xTitle='Year',yTitle='population',title='year vs population') 
```

# Barplot


```python
df2020 = df[df.index.isin([2020])]
df2020 = df2020.T
```

```python
df2020.iplot(kind='bar',xTitle='conuntries',yTitle='population',title=' population 2020', color='white' ) 
```

```python
dfdecada = df[df.index.isin([1980,1990,2000,2010,2020])]
dfdecada.iplot(kind='bar',xTitle='conuntries',yTitle='population',title=' population 2020' ) 
```

# Boxplot


```python
df.iplot(kind='box')
```

# Histograma

```python
df[['United States','China']].iplot(kind='hist',title='population United States')
```

# piechart 

```python
df2020 = df2020.reset_index()
df2020 = df2020.rename(columns={2020:'2020' })
df2020
```

```python
df2020.iplot(kind ='pie',labels='country',values='2020',title= 'population comparation')
```

# scatter plot

```python
df.iplot(kind='scatter',mode='markers',xTitle='Year',yTitle='population',title= 'Year vs population')
```
